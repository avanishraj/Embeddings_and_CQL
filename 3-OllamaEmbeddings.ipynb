```python
from langchain_community.embeddings import OllamaEmbeddings
embeddings = (
    OllamaEmbeddings(model="gemma:2b") #by default it uses llama2
) 
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "My name is Avanish Raj Singh",
        "My brother name is Prince"
    ]
)
r1
```

```python
print(r1[0])
```

```python
r2 = embeddings.embed_query(text="Who is Avanish Raj Singh")
r2
```
